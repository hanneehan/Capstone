```python
import pandas as pd
import numpy as np
import json
import joblib
import time
import matplotlib.pyplot as plt
```

```python
df_user = pd.read_json('yelp_academic_dataset_user.json', lines=True, nrows = 10)
df_user.columns
```

```python
df_user.head()
```

```python
# The reading of data is timed for experiment. 
start_time = time.time()

# Create empty lists to store variables of original dataset
u_id = []
name =[]
review_count = []
yelp_since = []
useful = []
funny = []
cool = []
elite = []
fan = []
avg_star = []
c_hot =[]
c_more =[]
c_profile =[]
c_cute =[]
c_list =[]
c_note =[]
c_plain =[]
c_cool =[]
c_funny =[]
c_writer =[]
c_photos =[]

#r = 0 # togerther with the muted code below was used for experimenting data reading time
with open("yelp_academic_dataset_user.json", "r") as f:
    for line in f.readlines():
        review = json.loads(line) # current line of the original data that has been read
        u_id.append(review['user_id'])
        name.append(review['name'])
        review_count.append(review['review_count'])
        yelp_since.append(review['yelping_since'])
        useful.append(review['useful'])
        funny.append(review['funny'])
        cool.append(review['cool'])
        elite.append(review['elite'])
        fan.append(review['fans'])
        avg_star.append(review['average_stars'])
        c_hot.append(review['compliment_hot'])
        c_more.append(review['compliment_more'])
        c_profile.append(review['compliment_profile'])
        c_cute.append(review['compliment_cute'])
        c_list.append(review['compliment_list'])
        c_note.append(review['compliment_note'])
        c_plain.append(review['compliment_plain'])
        c_cool.append(review['compliment_cool'])
        c_funny.append(review['compliment_funny'])
        c_writer.append(review['compliment_writer'])
        c_photos.append(review['compliment_photos'])
        
      
        #r += 1
        #if r > 1000:
            #break

end_time = time.time()
end_time - start_time  # Reading the entire dataset except text variable took 23s.
```

```python
# Create dataframe using the lists of variables
df_user = pd.DataFrame(list(zip(u_id, name, review_count, yelp_since, useful, funny, cool, elite, fan, avg_star,
                               c_hot, c_more, c_profile, c_cute, c_list, c_note, c_plain, c_cool, c_funny, c_writer,
                               c_photos)), 
                        columns = ['u_id', 'name', 'review_count', 'yelp_since', 'useful', 'funny', 'cool', 'elite', 'fan', 'avg_star',
                               'c_hot', 'c_more', 'c_profile', 'c_cute', 'c_list', 'c_note', 'c_plain', 'c_cool', 'c_funny', 'c_writer',
                               'c_photos'])
print(df_user.shape) # nrow = 1987897 , ncol = 21
df_user.head()
```

```python
# Dump the dataframe as 'review_no_text.py' in the local working directory
joblib.dump(df_user, 'user_no_friend.py') # 433M compared to 3G, much smaller
```
